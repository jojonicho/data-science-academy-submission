```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from Ipython.display import display
%matplotlib inline
```

```python
file1 = "catatan_lokasi.csv"
file2 = "data_profil.csv"
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
df = pd.merge(df1,df2, on="id")
cols = ['id', 'jenis_kelamin', 'divisi', 'umur', 'tanggal','lokasi_dominan']
df = df[cols]
df.lokasi_dominan.value_counts()
```
